```python
from PIL import Image
from numpy import asarray
import numpy as np
```

```python

# Read all the images

img1 = Image.open(' wp4411453.jpg')
img2 = Image.open('388294.jpg')
img3 = Image.open('HD-wallpaper-black-panther-black-panther-comics-marvel-mcu-wakanda.jpg')
img4 = Image.open('HD-wallpaper-tom-and-jerry-happy-tom-and-jerry-cartoon-animation-thumbnail.jpg')
img5 = Image.open('cdd98af2d4da5c6b587f331ee62deb6b.jpg')
img6 = Image.open('HD-wallpaper-dexter-cartoon-idk.jpg')
img7 = Image.open('HD-wallpaper-shinchan-anime-cartoon.jpg')
img8 = Image.open('loki-comic-art-4k-bn.jpg')
img9 = Image.open('wp6340297.jpg')
img10 = Image.open('281180.jpg')

#########################################################################################



#########################################################################################


#Resize all images to a standard height and width. Here I have used rs as resize notation

rs1 = img1.resize((200,200))
rs2 = img2.resize((200,200))
rs3 = img3.resize((200,200))
rs4 = img4.resize((200,200))
rs5 = img5.resize((200,200))
rs6 = img6.resize((200,200))
rs7 = img7.resize((200,200))
rs8 = img8.resize((200,200))
rs9 = img9.resize((200,200))
rs10 = img10.resize((200,200))

#########################################################################################

# Convert all resized images to array using asarray() numpy function
d1 = asarray(rs1)
d2 = asarray(rs2)
d3 = asarray(rs3)
d4 = asarray(rs4)
d5 = asarray(rs5)
d6 = asarray(rs6)
d7 = asarray(rs7)
d8 = asarray(rs8)
d9 = asarray(rs9)
d10 = asarray(rs10)

#########################################################################################


#########################################################################################

'''
Now using hstack() we can stack the matrices in horizontal order. hstack() is numpy function.

hstack() goes on appending matrices passed to it. It works similarly like append() function.
Here, it will go on appending the images(matrix) horizontally.

'''
horizontal_array1 = np.hstack([d1,d2,d3,d4,d5])

horizontal_array2 = np.hstack([d6,d7,d8,d9,d10])


'''
vstack goes on appending matrices passed to it. It works similarly like append() function.
Here, it will go on appending the images(matrix) vertically.
'''
vertical_array=np.vstack([horizontal_array1,horizontal_array2])

#########################################################################################

# Finally convert the matrix array back to image form using fromarray() function of PIL(Pillow) library.
image2 = Image.fromarray(vertical_array.astype('uint8'),'RGB')
image2.save('Image_Collage.png')
image2.show()

```
